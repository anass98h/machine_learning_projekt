```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
```

#  Clustering

```python

feature_df = pd.read_excel("ML/data/classifier/AimoScore_WeakLink_big_scores.xls")
#df = feature_df.drop(columns = ["ID", "AimoScore", "EstimatedScore"])
weaklink_df = pd.read_excel("ML/data/classifier/20190108 scores_and_weak_links.xlsx")
df = weaklink_df.drop(columns = ["ID", "Date", "SCORE"])
# weaklink_df.head()
#df = feature_df.merge(weaklink_df, on="ID", how="inner")
#df.drop(columns=["ID"], inplace=True)
print(df.columns)
df.head()
```

```python
from sklearn.model_selection import train_test_split

kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(df)
print("Cluster labels for each observation:")
df["cluster"] = kmeans.labels_
df.head(100)
```

# PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

featture_used = df.drop(columns = ["cluster"])
pcs = PCA(n_components=2)
transformed_data = pcs.fit_transform(featture_used)
explained_variance = pcs.explained_variance_
explained_variance_ratio = pcs.explained_variance_ratio_

print(df.shape)
print(transformed_data.shape)

# Create loadings dataframe
loadings = pd.DataFrame(pcs.components_.T, index=featture_used.columns, columns=[f'Comp.{i+1}' for i in range(pcs.n_components_)])

# Calculate sum of squared loadings (SS loadings)
ss_loadings = np.sum(loadings**2, axis=0)

# Create a comprehensive summary table
print(" Loadings:")
print(loadings)
print("\n SS loadings")
print(pd.Series(ss_loadings, index=[f'Comp.{i+1}' for i in range(pcs.n_components_)]))
print("Proportion Var")
print(pd.Series(explained_variance_ratio, index=[f'Comp.{i+1}' for i in range(pcs.n_components_)]))
print("Cumulative Var")
print(pd.Series(np.cumsum(explained_variance_ratio), index=[f'Comp.{i+1}' for i in range(pcs.n_components_)]))

#Create visualization
plot_df = pd.DataFrame({
    'PC1': transformed_data[:, 0],
    'PC2': transformed_data[:, 1],
    'Cluster': df['cluster']
})

# Plot results
plt.figure(figsize=(10, 8))
sns.scatterplot(data=plot_df, x='PC1', y='PC2', hue='Cluster', palette='viridis')
plt.title('Clusters Visualized in PCA Space')
plt.show()
```

```python



# Create a DataFrame with the PCA scores and cluster labels
pca_df = pd.DataFrame(
    data=transformed_data,
    columns=['PC1', 'PC2']
)

# Reattach the cluster labels -make sure this matches the original data
pca_df['Cluster'] = kmeans.labels_  # Make sure this variable exists from earlier

# Create the plot
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='PC1', 
    y='PC2', 
    hue='Cluster',
    palette='viridis',
    s=100,  # Point size
    alpha=0.7,  # Transparency
    data=pca_df
)

# Add loading vectors (biplot)
# Scale the loadings for better visibility
scale = 4  # Adjust this scale as needed for visibility
for i, feature in enumerate(featture_used.columns):
    plt.arrow(
        0, 0,  # Start at origin
        loadings.iloc[i, 0] * scale,  # PC1 loading
        loadings.iloc[i, 1] * scale,  # PC2 loading
        head_width=0.1,
        head_length=0.1,
        fc='red',
        ec='red'
    )
    plt.text(
        loadings.iloc[i, 0] * scale * 1.15,  # Slightly further than the arrow
        loadings.iloc[i, 1] * scale * 1.15,
        feature,
        color='black',
        ha='center',
        va='center'
    )

# Add explained variance to axis labels
plt.xlabel(f'Principal Component 1 ({explained_variance_ratio[0]:.1%} variance)')
plt.ylabel(f'Principal Component 2 ({explained_variance_ratio[1]:.1%} variance)')

# Add title and grid
plt.title('PCA Score Plot with Cluster Assignments', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.7)

# Add legend with a title
plt.legend(title='Cluster')

# Add text about cumulative variance
plt.annotate(
    f'Cumulative explained variance: {np.sum(explained_variance_ratio):.1%}',
    xy=(0.02, 0.02),
    xycoords='axes fraction',
    fontsize=12
)

plt.tight_layout()
plt.show()
```

# Interpretation of PCA Results

PCA results provide valuable insights into the structure of the dataset, which appears to consist of various body movement or posture measurements. Let me break down what these results mean:

## Loadings

**Interpretation of PC1**: This component primarily captures forward-leaning upper body posture (arms falling forward, head forward) contrasted with outward knee movement. This could represent a "forward collapse vs. outward stance" dimension.

### Principal Component 2 (20.1% of variance):

## Practical Implications

1. **Data Reduction**: These two components can serve as a simplified representation of the original data, reducing 14 variables to 2 while retaining about 47% of information.

2. **Feature Grouping**: The loadings reveal which variables tend to co-occur:
   - Forward head and forward arm postures tend to happen together
   - Shoulder elevations tend to occur together
   - Knee movements show patterns related to overall posture

3. **For Classification/Clustering**: If we're using this for identifying posture types or movement patterns, these components suggest that:
   - The primary distinguishing factor is forward-leaning upper body posture
   - The secondary factor relates to shoulder elevation patterns

4. **Potential Redundancy**: Variables with high loadings in the same direction on the same component may be somewhat redundant. For example, left and right shoulder elevation contribute similarly to PC2.
