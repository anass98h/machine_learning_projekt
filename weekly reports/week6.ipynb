## 1. Improving ensembles

### 1.1 Classification ensemble

We added three different tree-based models to our ensemble. 

Our ensemble now consists of five models:

* KNN
* Logistic Regression
* **Bagging Tree**
* **Random Forest**
* **Boosting Tree**

These are the results:

- **Voting Ensemble**:
  - CV Mean Accuracy: 0.945437407018509  
  - CV Std Accuracy: 0.006907042204880505
  - Test Accuracy: 0.9518140589569161
- **Stacking Ensemble:**
  - CV Mean Accuracy: 0.9638608424700301 
  - CV Std Accuracy: 0.004411976272010811
  - Test Accuracy: **0.9767573696145124**



**Observation:**
Both ensemble methods were significantly improved by including the tree-based models. The new stacking ensemble is now our strongest classifier model with an accuracy of 97.6%.
